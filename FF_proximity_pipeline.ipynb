```python
from geopy.distance import geodesic
```

```python
mc_data = pd.read_excel("McDonalds.xlsx")
subway_data = pd.read_csv("subway.csv")
```

```python
def FF_calc(loc):
    # Calculates number of FFs in 0.1, 0.25, 0.5 and 1 miles of the loc
    ans = [0, 0, 0, 0]

    for i in range(len(mc_data)):
        mc_loc = (mc_data.iloc[i]["Latitude"], mc_data.iloc[i]["Longitude"])
        dist = geodesic(loc, mc_loc).miles

        if dist < 1:
            ans[3] += 1
            if dist < 0.5:
                ans[2] += 1
                if dist < 0.25:
                    ans[1] += 1
                    if dist < 0.1:
                        ans[0] += 1

    for i in range(len(subway_data)):
        subway_loc = (subway_data.iloc[i]["latitude"], subway_data.iloc[i]["longitude"])
        dist = geodesic(loc, subway_loc).miles

        if dist < 1:
            ans[3] += 1
            if dist < 0.5:
                ans[2] += 1
                if dist < 0.25:
                    ans[1] += 1
                    if dist < 0.1:
                        ans[0] += 1
        
    return ans
```
