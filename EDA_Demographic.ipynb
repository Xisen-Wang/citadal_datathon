```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
DATA_PATH = os.getcwd()
dataset_name = "Official Demographic Data Key Questions.xlsx"
FULL_PATH = os.path.join(DATA_PATH, 'data', dataset_name)
data = pd.read_excel(FULL_PATH, sheet_name='Plotting_v2')
data = data.head()

# Example 1: Histogram for 'All'
plt.figure(figsize=(10, 6))
sns.histplot(data['All'], kde=True, color='blue')
plt.title('Distribution of Responses for All Respondents')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()

# Example 2: Bar Plot for 'Male' and 'Female'
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(data))
plt.bar(index, data['Male'], width=bar_width, label='Male', color='b')
plt.bar([p + bar_width for p in index], data['Female'], width=bar_width, label='Female', color='r')
plt.xlabel('Question')
plt.ylabel('Percentage')
plt.title('Comparison of Male and Female Responses')
plt.xticks([p + bar_width / 2 for p in index], data['%'])
plt.legend()
plt.show()

# Example 3: Error Bar Plot for 'Asian' with Confidence Intervals
plt.figure(figsize=(10, 6))
plt.errorbar(data['%'], data['Asian'], yerr=[data['Asian'] - data['Asian CI low'], data['Asian CI high'] - data['Asian']],
             fmt='o', ecolor='g', capthick=2, label='Confidence Interval')
plt.title('Asian Responses with Confidence Intervals')
plt.xlabel('Question')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

```python
import os
import pandas as pd

# Define the base path and the dataset name
DATA_PATH = os.getcwd()
dataset_name = "Official Demographic Data Key Questions.xlsx"
# Construct the full path to the dataset
FULL_PATH = os.path.join(DATA_PATH, 'data', dataset_name)

# Load the data from the specific sheet named "Plotting_v2"
data = pd.read_excel(FULL_PATH, sheet_name='Plotting_v2')

# Assuming the data has non-numeric rows within the dataset such as "All CI low"
# Print the data to manually check for such rows
print(data.head(10))

# Remove non-numeric rows: adjust the row index as needed based on your actual data inspection
data = data[pd.to_numeric(data['All'], errors='coerce').notna()]

# Now convert columns to numeric, this will turn non-convertible values to NaN automatically
numeric_columns = ['All', 'Male', 'Female', 'Asian', 'Hispanic', 'Non-Hispanic Black', 'Non-Hispanic White',
                   '2 or more races', 'American Indian/Alaska Native', 'Hawaiian/Pacific Islander CI high', 'Other Races CI high',
                   'Male CI high', 'Female CI high', 'Asian CI high', 'Hispanic CI high', 'Non-Hispanic Black CI high', 'Non-Hispanic White CI high',
                   '2 or more races CI high', 'American Indian/Alaska Native CI high']  # Add other numeric columns as needed

for column in numeric_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')

# Now your data should be clean and only contain numeric values in the specified columns
print(data.head())

# Continue with further analysis or plotting
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Comparative bar plot for 'Male' and 'Female'
plt.figure(figsize=(12, 6))
sns.barplot(x='%', y='Male', data=data, color='blue', label='Male')
sns.barplot(x='%', y='Female', data=data, color='red', alpha=0.6, label='Female')
plt.title('Comparison of Male and Female Responses Across Questions')
plt.xlabel('Questions')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

```python
# Box plot to show distributions for different demographics
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[['Asian', 'Hispanic', 'Non-Hispanic Black', 'Non-Hispanic White']])
plt.title('Distribution of Responses Across Ethnic Groups')
plt.xlabel('Ethnic Groups')
plt.ylabel('Percentage')
plt.show()
```

```python
# Stacked bar plot for responses (assuming categorical 'Agree', 'Disagree', 'Neutral')
data['Agree'] = data['All'] * 0.5  # Example data manipulation
data['Disagree'] = data['All'] * 0.3
data['Neutral'] = data['All'] * 0.2

plt.figure(figsize=(12, 6))
data.set_index('%')[['Agree', 'Disagree', 'Neutral']].plot(kind='bar', stacked=True, color=['green', 'red', 'gray'])
plt.title('Stacked Bar Plot of Responses')
plt.xlabel('Questions')
plt.ylabel('Percentage')
plt.legend(title='Response Type')
plt.show()
```

```python
# Error bar plot showing responses with their confidence intervals for 'Asian'
plt.figure(figsize=(12, 6))
plt.errorbar(x=data['%'], y=data['Asian'], yerr=[data['Asian'] - data['Asian CI low'], data['Asian CI high'] - data['Asian']],
             fmt='o', ecolor='r', capsize=5)
plt.title('Asian Responses with Confidence Intervals')
plt.xlabel('Question')
plt.ylabel('Percentage')
plt.show()
```

```python
# Heatmap showing correlations between different demographics
plt.figure(figsize=(10, 8))
sns.heatmap(data[['Male', 'Female', 'Asian', 'Hispanic']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Between Demographic Groups')
plt.show()
```

```python
# Density plot for 'Male' and 'Female'
plt.figure(figsize=(12, 6))
sns.kdeplot(data['Male'], shade=True, color='blue', label='Male')
sns.kdeplot(data['Female'], shade=True, color='red', label='Female')
plt.title('Density Plot of Male and Female Responses')
plt.xlabel('Percentage')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Scatter plot comparing 'Male' vs 'Female'
plt.figure(figsize=(10, 6))
plt.scatter(data['Male'], data['Female'], alpha=0.5)
plt.title('Scatter Plot Comparing Male and Female Responses')
plt.xlabel('Male Percentage')
plt.ylabel('Female Percentage')
plt.show()
```

```python
# Facet Grid using Seaborn for 'Male' and 'Female' across multiple questions
g = sns.FacetGrid(data, col='%', col_wrap=4, height=3)
g.map(sns.barplot, 'Male', 'All')
plt.show()
```

```python
# Pie chart for demographic responses (taking first row as an example)
labels = ['Male', 'Female']
sizes = [data.loc[0, 'Male'], data.loc[0, 'Female']]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Response Distribution for Question 1')
plt.show()
```

```python
import os
import pandas as pd

# Define the base path and the dataset name
DATA_PATH = os.getcwd()
dataset_name = "Official Demographic Data Key Questions.xlsx"
# Construct the full path to the dataset
FULL_PATH = os.path.join(DATA_PATH, 'data', dataset_name)

# Load the data from the specific sheet named "plotting"
# and skip the first row using skiprows parameter
data = pd.read_excel(FULL_PATH, sheet_name='Plotting_rotated')

# Print the data
print(data)
```

### Checking Correlations

```python
# Calculate correlation matrix
correlation_matrix = data[['Q036', 'Q037', 'Q038', 'Q039', 'Q058']].corr()
print("Correlation Matrix:\n", correlation_matrix)

# Plotting the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Obesity, Overweight, and Soda Consumption')
plt.show()
```

```python
# Plotting obesity rates for different demographics
plt.figure(figsize=(12, 6))
sns.barplot(x='%', y='Q036', data=data)
plt.title('Obesity Rates Across Demographics')
plt.xlabel('Demographic Groups')
plt.ylabel('Obesity Rate')
plt.xticks(rotation=45)
plt.show()
```

### Investigation On the Questions

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example data preparation
# Extract main data and corresponding confidence intervals
main_data = data[data['%'].isin(['All', 'Male', 'Female', 'Asian', 'Hispanic', 'Non-Hispanic Black',
                                 'Non-Hispanic White', '2 or more races', 'American Indian/Alaska Native',
                                 'Hawaiian/Pacific Islander', 'Other Races'])]

# Lower CI
lower_ci = data[data['%'].isin(['All CI low', 'Male CI low', 'Female CI low', 'Asian CI low', 'Hispanic CI low',
                                'Non-Hispanic Black CI low', 'Non-Hispanic White CI low', '2 or more races CI low',
                                'American Indian/Alaska Native CI low', 'Hawaiian/Pacific Islander CI low', 'Other Races CI low'])]

# Upper CI
upper_ci = data[data['%'].isin(['All CI high', 'Male CI high', 'Female CI high', 'Asian CI high', 'Hispanic CI high',
                                'Non-Hispanic Black CI high', 'Non-Hispanic White CI high', '2 or more races CI high',
                                'American Indian/Alaska Native CI high', 'Hawaiian/Pacific Islander CI high', 'Other Races CI high'])]

# Normalize the index for easy subtraction
lower_ci.index = main_data.index
upper_ci.index = main_data.index

# Calculate errors (distance from main data to confidence bounds)
errors = upper_ci[['Q036']] - lower_ci[['Q036']]

# Plotting with error bars
plt.figure(figsize=(12, 6))
plt.bar(main_data['%'], main_data['Q036'], yerr=errors['Q036'].values.T, capsize=5, color='skyblue')
plt.title('Obesity Rates of Adults Across Demographics with 99% Confidence Intervals')
plt.xlabel('Demographic Groups')
plt.ylabel('Obesity Rate')
plt.xticks(rotation=45)
plt.show()
```
