## This is an EDA of the 2022 Overall prevalence dataset

```python
# Import all the necessary packages
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Plot the Prevalence of Obesity Across States

```python
# Load the dataset
DATA_PATH = os.getcwd()
dataset_name = "2022-overall-prevalence.csv"
FULL_PATH = os.path.join(DATA_PATH, 'data', dataset_name)
data = pd.read_csv(FULL_PATH)
data.head()
```

```python
# Plotting

# Process the '95% CI' column to split it into 'CI Lower' and 'CI Upper'
data[['CI Lower', 'CI Upper']] = data['95% CI'].str.extract('\((.*), (.*)\)').astype(float)

# Set up the plotting area
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Create a scatter plot with error bars
plt.errorbar(data['State'], data['Prevalence'], 
             yerr=[data['Prevalence'] - data['CI Lower'], data['CI Upper'] - data['Prevalence']],
             fmt='o', ecolor='gray', capsize=3, linestyle='None', marker='s', color='blue')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Prevalence (%)')
plt.title('Prevalence of Obesity by State with 95% CI')

# Rotate state labels for better visibility if necessary
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
plt.savefig()
```

```python
# Process the '95% CI' column to split it into 'CI Lower' and 'CI Upper'
data[['CI Lower', 'CI Upper']] = data['95% CI'].str.extract('\((.*), (.*)\)').astype(float)

# Sort data by 'Prevalence'
data.sort_values('Prevalence', inplace=True)

# Set up the plotting area
plt.figure(figsize=(12, 8))  # Increased figure size for better visibility
sns.set(style="whitegrid")

# Create a scatter plot with error bars
plt.errorbar(data['State'], data['Prevalence'], 
             yerr=[data['Prevalence'] - data['CI Lower'], data['CI Upper'] - data['Prevalence']],
             fmt='o', ecolor='gray', capsize=3, linestyle='None', marker='s', color='blue')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Prevalence (%)')
plt.title('Prevalence of Condition by State with 95% CI, Sorted by Prevalence')

# Rotate state labels for better visibility
plt.xticks(rotation=90)  # Rotate labels to 90 degrees to avoid overlap

# Show the plot
plt.tight_layout()
plt.show()
```
